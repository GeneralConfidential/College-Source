```python
import pandas as pd
import numpy as np
import sklearn.linear_model as linear_model
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("Startups.csv")
```

```python
plt.xlabel("R&D")
plt.ylabel("Profit")
plt.scatter(df.rnd, df.prof)
reg1 = linear_model.LinearRegression()
reg1.fit(df[["rnd"]], df.prof)
plt.plot(df.rnd, reg1.predict(df[["rnd"]]), color='cyan')
```

```python
plt.xlabel("Admin")
plt.ylabel("Profit")
plt.scatter(df.admin, df.prof)
reg1 = linear_model.LinearRegression()
reg1.fit(df[["admin"]], df.prof)
plt.plot(df.admin, reg1.predict(df[["admin"]]), color='yellow')
```

```python
plt.xlabel("Mkt")
plt.ylabel("Profit")
plt.scatter(df.mkt, df.prof)
reg1 = linear_model.LinearRegression()
reg1.fit(df[["mkt"]], df.prof)
plt.plot(df.mkt, reg1.predict(df[["mkt"]]), color='magenta')
```
